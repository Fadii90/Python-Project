```python
import pandas as pd # data processing 
import numpy as np  #linear algebra (works with array )
import matplotlib.pyplot as plt # plots results
import seaborn as sns
from sklearn.model_selection import train_test_split  #divide dataset
from sklearn.linear_model import LinearRegression    #implement linearregression 


#import our data (data collection )
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/advertising.csv")
# view first 5 entries
print(data.head())
```

```python
data.shape
```

```python
#data preparation 
print(data.isnull().sum())
```

```python
#corellations btwn features
#feature extaraction 
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr())
plt.show()
```

```python
#relationship between the independent variables and the dependent variable.
#use a linear regression algorithm to train a sales prediction model  using tv and sales 
#So we will use just the TV feature to build our simple linear regression model since it has the highest correlation with Sales.



#Creating two arrays one is for containing lable and other one is containing features(1 dim and multidimen)
x = np.array(data.drop(["Sales"], 1))
y = np.array(data["Sales"])

# test the data and train the data 
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)

#create an object of the LinearRegression class and fit it to our training data using the fit() method.
model = LinearRegression()
model.fit(xtrain, ytrain)

#We will now evaluate our model on the test set. 
ypred = model.predict(xtest)


data = pd.DataFrame(data={"Predicted Sales": ypred.flatten()})
print(data)
```
